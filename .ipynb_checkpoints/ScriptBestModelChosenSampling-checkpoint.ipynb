```python
numeriƒçke_varijable=['Age', 'DailyRate', 'DistanceFromHome', 'HourlyRate', 'MonthlyIncome',
                     'MonthlyRate', 'NumCompaniesWorked', 'PercentSalaryHike', 'TotalWorkingYears',
                    'TrainingTimesLastYear', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 
                    'YearsSinceLastPromotion']
```

```python
kategoriƒçke_nominalne=['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']
```

```python
izlazna_varijabla=['Attrition']
```

```python
kategoriƒçke_binarne=['Gender', 'OverTime']
```

```python
kategoriƒçke_ordinalne=['Education', 'EnvironmentSatisfaction', 'JobInvolvement', 
                       'JobLevel', 'JobSatisfaction', 'PerformanceRating', 'RelationshipSatisfaction',
                       'StockOptionLevel', 'WorkLifeBalance']
```

```python
###### Ispod je predstavljena tehnika otklanjanja outlier-a na osnovu z-scorea
```

```python
data_attrition.shape
```

```python
###
import numpy as np
from scipy import stats
```

```python
z_skorovi = np.abs(stats.zscore(data_attrition[numeriƒçke_varijable]))
data_attrition_outlier_cleaned = data_attrition[(z_skorovi < 3).all(axis=1)]
```

```python
data_attrition_outlier_cleaned.shape
```

```python
#### Sledeƒái korak jeste feature engineering u kome ƒáemo dodati odredjene izvedene varijable
```

```python
data_attrition_obogacen= data_attrition_outlier_cleaned.copy()
```

```python
data_attrition_obogacen['YearsAtCompany_ManagerRatio'] = data_attrition_obogacen['YearsAtCompany'] / (data_attrition_obogacen['YearsWithCurrManager'] + 1)
data_attrition_obogacen['SalaryIncreasePerYear']=(data_attrition_obogacen['MonthlyIncome']-data_attrition_obogacen['DailyRate']*22) / data_attrition_obogacen['YearsAtCompany']
```

```python
data_attrition_obogacen.shape
```

```python
### sada se dodaju ove dve numericke varijable u numericke varijable i onda je potrebno da ƒáe biti odradjena 
##odgovarajuƒáa konverzija kategorickih varijabli u numerickih kao i skaliranje numerickih varijabli
```

```python
numeriƒçke_varijable.append('YearsAtCompany_ManagerRatio')
numeriƒçke_varijable.append('SalaryIncreasePerYear')
```

```python
numeriƒçke_varijable
```

```python
##### konverzija kategorickih u numericke
```

```python
import warnings

warnings.filterwarnings('ignore', category=FutureWarning, message='is_sparse is deprecated')
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
label_encoders = {}
for kolona in kategoriƒçke_ordinalne:
    le = LabelEncoder()
    data_attrition_obogacen[kolona] = le.fit_transform(data_attrition_obogacen[kolona])
    label_encoders[kolona] = le
```

```python
data_attrition_obogacen = pd.get_dummies(data_attrition_obogacen, columns= kategoriƒçke_nominalne + kategoriƒçke_binarne, drop_first=True)
```

```python
#### Sada se koristi Standard Scaler za skaliranje numerickih varijabli
```

```python
###3 ! ovo je zbog izvedenih varijabli 
data_attrition_obogacen.replace([np.inf, -np.inf], np.nan, inplace=True)
data_attrition_obogacen.dropna(inplace=True)
```

```python
data_attrition_obogacen.shape
```

```python
scaler = StandardScaler()
data_attrition_obogacen[numeriƒçke_varijable] = scaler.fit_transform(data_attrition_obogacen[numeriƒçke_varijable])
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=30)
data_pca = pca.fit_transform(data_attrition_obogacen.drop(columns=['Attrition']))
data_pca = pd.DataFrame(data_pca, columns=[f'PC{i}' for i in range(1, 31)])
data_pca['Attrition'] = data_attrition_obogacen['Attrition'].values
```

```python
data_pca
```

```python
### podela skupa podataka na treniranje i testiranje
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_ulazi = data_pca.drop(columns=['Attrition'])
y_izlaz = data_pca['Attrition']
X_train, X_test, y_train, y_test = train_test_split(X_ulazi, y_izlaz, test_size=0.2, random_state=42, stratify=y_izlaz)
```

```python
######### obrizom na ne balansiranost klasa potrebno je uraditi i balansking samih klasa
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
import xgboost as xgb
import lightgbm as lgb
import catboost as cb
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, roc_auc_score, roc_curve, precision_score, recall_score, f1_score, precision_recall_curve
```

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'Bagging': BaggingClassifier(),
    'XGBoost': xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    'LightGBM': lgb.LGBMClassifier(),
    'CatBoost': cb.CatBoostClassifier(silent=True)
}
```

```python
results_no_tuning = {}
for model_name, model in models.items():
    print(f"Training {model_name} without hyperparameter tuning...")
    model.fit(X_train_bal, y_train_bal)
    y_pred = model.predict(X_test)
    y_pred_prob = model.predict_proba(X_test)[:, 1]
    
    results_no_tuning[model_name] = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1 Score': f1_score(y_test, y_pred),
        'ROC AUC': roc_auc_score(y_test, y_pred_prob),
        'Confusion Matrix': confusion_matrix(y_test, y_pred),
        'Classification Report': classification_report(y_test, y_pred)
    }
```

```python
for model_name, result in results_no_tuning.items():
    print(f"Model: {model_name}")
    print(f"Accuracy: {result['Accuracy']}")
    print(f"Precision: {result['Precision']}")
    print(f"Recall: {result['Recall']}")
    print(f"F1 Score: {result['F1 Score']}")
    print(f"ROC AUC: {result['ROC AUC']}")
    print(f"Confusion Matrix:\n{result['Confusion Matrix']}")
    print(f"Classification Report:\n{result['Classification Report']}\n")

    # Plot Confusion Matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(result['Confusion Matrix'], annot=True, fmt='d', cmap='Blues', xticklabels=['No Attrition', 'Attrition'], yticklabels=['No Attrition', 'Attrition'])
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'Confusion Matrix for {model_name} (No Tuning)')
    plt.show()
    
    # Plot ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {result["ROC AUC"]:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve for {model_name} (No Tuning)')
    plt.legend()
    plt.show()
    
    # Plot Precision-Recall curve
    precision, recall, _ = precision_recall_curve(y_test, y_pred_prob)
    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, label=f'{model_name}')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(f'Precision-Recall Curve for {model_name} (No Tuning)')
    plt.legend()
    plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grids = {
    'Logistic Regression': {
        'C': [0.01, 0.1, 1, 10, 100],
        'solver': ['lbfgs', 'liblinear']
    },
    'Decision Tree': {
        'max_depth': [None, 10, 20, 30, 40],
        'min_samples_split': [2, 5, 10]
    },
    'Random Forest': {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10]
    },
    'Gradient Boosting': {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 5, 7]
    },
    'AdaBoost': {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.01, 0.1, 1]
    },
    'Bagging': {
        'n_estimators': [10, 50, 100],
        'max_samples': [0.5, 1.0],
        'max_features': [0.5, 1.0]
    },
    'XGBoost': {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 5, 7]
    },
    'LightGBM': {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.01, 0.1, 0.2],
        'num_leaves': [31, 62, 127]
    },
    'CatBoost': {
        'iterations': [100, 200, 300],
        'learning_rate': [0.01, 0.1, 0.2],
        'depth': [3, 5, 7]
    }
}

# Train, tune, and evaluate models with hyperparameter tuning
results_with_tuning = {}
for model_name, model in models.items():
    print(f"Training {model_name} with hyperparameter tuning...")
    grid_search = GridSearchCV(model, param_grids[model_name], cv=3, scoring='accuracy')
    grid_search.fit(X_train_bal, y_train_bal)
    
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test)
    y_pred_prob = best_model.predict_proba(X_test)[:, 1]
    
    results_with_tuning[model_name] = {
        'Best Params': grid_search.best_params_,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1 Score': f1_score(y_test, y_pred),
        'ROC AUC': roc_auc_score(y_test, y_pred_prob),
        'Confusion Matrix': confusion_matrix(y_test, y_pred),
        'Classification Report': classification_report(y_test, y_pred)
    }

    # Display results for models with hyperparameter tuning
    print(f"Model: {model_name}")
    print(f"Best Params: {results_with_tuning[model_name]['Best Params']}")
    print(f"Accuracy: {results_with_tuning[model_name]['Accuracy']}")
    print(f"Precision: {results_with_tuning[model_name]['Precision']}")
    print(f"Recall: {results_with_tuning[model_name]['Recall']}")
    print(f"F1 Score: {results_with_tuning[model_name]['F1 Score']}")
    print(f"ROC AUC: {results_with_tuning[model_name]['ROC AUC']}")
    print(f"Confusion Matrix:\n{results_with_tuning[model_name]['Confusion Matrix']}")
    print(f"Classification Report:\n{results_with_tuning[model_name]['Classification Report']}\n")

    # Plot Confusion Matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(results_with_tuning[model_name]['Confusion Matrix'], annot=True, fmt='d', cmap='Blues', xticklabels=['No Attrition', 'Attrition'], yticklabels=['No Attrition', 'Attrition'])
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'Confusion Matrix for {model_name} (With Tuning)')
    plt.show()
    
    # Plot ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {results_with_tuning[model_name]["ROC AUC"]:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve for {model_name} (With Tuning)')
    plt.legend()
    plt.show()
    
    # Plot Precision-Recall curve
    precision, recall, _ = precision_recall_curve(y_test, y_pred_prob)
    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, label=f'{model_name}')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(f'Precision-Recall Curve for {model_name} (With Tuning)')
    plt.legend()
    plt.show()
```
